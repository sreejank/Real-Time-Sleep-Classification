First, I'll try unsupervised spectral clustering on subject 1032 to see if it works better. To seperate Wake/Sleep, I will use EMG Power, 0.5-55 Hz/0.5-22 Hz EEG Power Ratio, and number of sign inversions.  

```python
%matplotlib inline
import numpy as np
import matplotlib as plt
import load,processing,clustering,plotting
from importlib import reload
eeg,emg,actual=load.getData('../Data/channels1032.dat','../Data/teststates1032.dat')
features=processing.getFeatures(eeg,emg)
```

```python
reload(plotting)
from sklearn.cluster import SpectralClustering
wakeFeatures=features[:,3:]
actualWake=[0 if state==0 else 1 for state in actual]
clusters=SpectralClustering(n_clusters=2)
wakeLabels=clusters.fit_predict(wakeFeatures)
plotting.plotFeatures(wakeFeatures,wakeLabels,actualWake,['EMG Power','Large/Small Ratio','Sign Inversions'])
```

This looks more decent. Let's see exactly how many of the states were correct.

```python
matches=0
for i in range(len(actualWake)):
    state1,state2=actualWake[i],wakeLabels[i]
    if state1==state2:
        matches+=1
print("Percent similarity between calculated and actual wake/sleep states: "+str(matches/len(actualWake)))
print()
clustering.printSimilarity(actualWake,wakeLabels)
```

Now let's see the performance in detecting REM sleep. Features used will be: Delta, Delta/Theta Ratio, and EMG Median

```python
sleepIndices=[i for i in range(len(wakeLabels)) if wakeLabels[i]==1]
#REM=1, NREM=0
actualREM=[1 if actual[i]==2 else 0 for i in sleepIndices]
remFeatures=[]
for i in sleepIndices:
    remFeatures.append(features[i,:3])
remFeatures=np.asarray(remFeatures)
clusters=SpectralClustering(n_clusters=2)
remLabels=clusters.fit_predict(remFeatures)
plotting.plotFeatures(remFeatures,remLabels,actualREM,['Delta','Delta/Theta Ratio','EMG Median'])
```

This doesn't look as good, but let's see the whole thing done in ensemble. 

```python
reload(clustering)
labels=clustering.doubleSpectralCluster(features)
plotting.plotFeatures(features,labels,actual,['Delta','Delta/Theta','EMG Med','EMG Amp','Large/Small','Sign Inv'])
```

```python
matches=0
for i in range(len(actual)):
    state1,state2=labels[i],actual[i]
    if state1==state2:
        matches+=1
print("Percent similarity between calculated and actual wake/sleep states: "+str(matches/len(actual)))
print()
clustering.printSimilarity(actual,labels)

remmatches=0
rem=0
for i in range(len(actual)):
    state1,state2=labels[i],actual[i]
    if state2==2:
        rem+=1
        if state1==state2:
            remmatches+=1
print("Percent similarity between calculated and actual rem states: "+str(remmatches/rem))
```

This does not hit as many rem states as kmeans, so spectral clustering did not work out as well. For now, I will start crossvalidating supervised to see how well that does.

```python
from sklearn.cross_validation import KFold
from sklearn import svm,tree
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
import time
reload(processing)
def supervisedCrossValidate(subjectnum):
    eeg,emg,actual=load.getData('../Data/channels{}.dat'.format(subjectnum),'../Data/teststates{}.dat'.format(subjectnum))
    features=processing.getFeatures(eeg,emg)
    features=np.asarray(features)
    actual=np.asarray(actual)
    
    clfsvm=svm.SVC(kernel='linear')
    clftree=tree.DecisionTreeClassifier(criterion='entropy',max_depth=4)
    clfnb=GaussianNB()
    clfknn=KNeighborsClassifier()
    print("Subject {}, Mean Classifier Accuracy, Mean REM Sensitivity, Mean REM Specificity, Mean NREM Sensitivity, Mean NREM Specificity".format(subjectnum))
    classifiers=[clfsvm,clftree,clfnb,clfknn]
    clfTime=[]
    for clf in classifiers:
        scores=[]
        sensitivitiesrem=[]
        specificitiesrem=[]
        sensitivitiesnrem=[]
        specificitiesnrem=[]
        print(clf.__class__.__name__,end=',')
        ctime=0
        kf=KFold(len(features),n_folds=4)
        for trainIndex,testIndex in kf:
            features_train,features_test=features[trainIndex],features[testIndex]
            labels_train,labels_test=actual[trainIndex],actual[testIndex]
            
            begin=time.time()
            clf.fit(features_train,labels_train)
            ctime+=time.time()-begin
            
            scores.append(clf.score(features_test,labels_test))
            calculatedLabels=clf.predict(features_test)
            
            remtp,remfp,remtn,remfn=0,0,0,0
            nremtp,nremfp,nremtn,nremfn=0,0,0,0
            
            for i in range(len(calculatedLabels)):
                if labels_test[i]==2:
                    if calculatedLabels[i]==2:
                        remtp+=1
                    else:
                        remfn+=1
                else:
                    if calculatedLabels[i]==2:
                        remfp+=1
                    else:
                        remtn+=1
                        
                if labels_test[i]==1:
                    if calculatedLabels[i]==1:
                        nremtp+=1
                    else:
                        nremfn+=1
                else:
                    if calculatedLabels[i]==1:
                        nremfp+=1
                    else:
                        nremtn+=1
                
            sensitivitiesrem.append(float(remtp)/(remtp+remfn))
            specificitiesrem.append(float(remtn)/(remtn+remfp))
            
            sensitivitiesnrem.append(float(nremtp)/(nremtp+nremfn))
            specificitiesnrem.append(float(nremtn)/(nremtn+nremfp))
            
        print((np.mean(scores)),end=',')
        print((np.mean(sensitivitiesrem)),end=',')
        print((np.mean(specificitiesrem)),end=',')
        print((np.mean(sensitivitiesnrem)),end=',')
        print((np.mean(specificitiesnrem)),end='')
        print()
        clfTime.append(ctime)
    
    print("-----------------")
    print("Classifier Time Array: ")
    print(clfTime)
    print("-----------------")
        

supervisedCrossValidate(1032)
supervisedCrossValidate(1039)
supervisedCrossValidate(1040)
supervisedCrossValidate(1134)
supervisedCrossValidate(1139)
supervisedCrossValidate(1140)

    
```

These are the statistics for Classifier Accuracy, REM Specificity/Sensitivity, and NREM Specificity/Sensitivity in CSV format, along with arrays of how long each classifier took (in order: SVM, Decision Tree, Naive Bayes, K Nearest Neighbors). Below is the graphical representation:
